### A1. Temperature converter  *(2 marks)*

Write a function `c_to_f(celsius)` that converts a temperature in Celsius to Fahrenheit
using the formula `F = C * 9/5 + 32`.

- It should work for a single number **and** for a list of numbers (return a list in that case).

```python
c_to_f(0)          # -> 32.0
c_to_f(100)        # -> 212.0
c_to_f([0, 37, 100])  # -> [32.0, 98.6, 212.0]
```

````python
# A1 — your solution here
def c_to_f(celsius):
    # If the input is a list, convert each element
    if isinstance(celsius, list): 
    #isinstance() checks if the input is a list
        
        new_list = [] 
        for c in celsius:
            new_list.append(c * 9/5 + 32)
            
        return new_list
         
    # For single number conversion
    else:
        return celsius * 9/5 + 32


choice = input("Enter 's' for single value or 'l' for list: ")
# Enters the user's choice as to convert a single data point or a number of such points in a list.
# The choice is converted to lowercase and then analyzed so as to determine the flow of the program.
if choice.lower() == 's':
    c = float(input("Enter temperature in Celsius: "))
    print("Temperature in Fahrenheit:", c_to_f(c))

elif choice.lower() == 'l':
    values = input("Enter Celsius values separated by spaces: ")
    c_list = [float(x) for x in values.split()]
    print("Temperatures in Fahrenheit:", c_to_f(c_list))

else:
    print("Invalid choice!")
````

### A2. FizzBuzz with a twist  *(3 marks)*

Write a function `fizzbuzz(n)` that returns a **list** of length `n` (covering numbers `1..n`) where:
- multiples of 3 are replaced by `"Fizz"`
- multiples of 5 are replaced by `"Buzz"`
- multiples of both 3 and 5 are replaced by `"FizzBuzz"`
- all other numbers stay as integers.

```python
fizzbuzz(5)  # -> [1, 2, 'Fizz', 4, 'Buzz']
```

````python
# A2 — your solution here
def fizzbuzz(n):
    result = []

    for i in range(1, n + 1):

        if i % 3 == 0 and i % 5 == 0:
            result.append("FizzBuzz")

        elif i % 3 == 0:
            result.append("Fizz")

        elif i % 5 == 0:
            result.append("Buzz")

        else:
            result.append(i)

    return result


# User input
n = int(input("Enter a number: "))

print(fizzbuzz(n))
````

### A3. Word frequency counter  *(4 marks)*

Write a function `word_count(text)` that takes a string and returns a **dictionary** mapping each
lowercase word to the number of times it appears. Ignore case and strip basic punctuation (`. , ! ?`).

```python
word_count("The cat sat. The cat ran!")
# -> {'the': 2, 'cat': 2, 'sat': 1, 'ran': 1}
```

````python
# A3 — your solution here
def word_count(text):

    # To convert the text to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.replace(".", "")
    text = text.replace(",", "")
    text = text.replace("!", "")
    text = text.replace("?", "")

    # Split the text into words
    words = text.split()

    # Create an empty dictionary
    count = {}

    # Count the occurrences of each word
    for i in words:
        if i in count:
            count[i] += 1
        else:
            count[i] = 1

    return count


# User input
text = input("Enter a sentence: ")

print(word_count(text))
````

### A4. Filter and transform  *(3 marks)*

containing the **squares of only the even numbers**, preserving order. Use a **list comprehension**.

```python
squares_of_evens([1, 2, 3, 4, 5, 6])  # -> [4, 16, 36]
```

````python
# A4 — your solution here
def squares_of_evens(nums):
    return [num ** 2 for num in nums if num % 2 == 0]


# User input
values = input("Enter integers separated by spaces: ")
nums = []
y=values.split()
for x in y:
    nums.append(int(x))

print("Squares of even numbers:", squares_of_evens(nums))
````

### A5. Running maximum  *(3 marks)*

Write a function `running_max(nums)` that returns a list where the element at index `i` is the
maximum of `nums[0..i]`. Solve it with a plain loop (no NumPy, no `max()` over slices).

```python
running_max([3, 1, 4, 1, 5, 9, 2, 6])  # -> [3, 3, 4, 4, 5, 9, 9, 9]
```

````python
# A5 — your solution here
def running_max(nums):

    result = []            
    # List to store running maximums
    current_max = nums[0]  
    # Assuming the first element to be the maximum

    for i in nums:
        if i > current_max:
            current_max = i

        result.append(current_max)

    return result


# User input
values = input("Enter integers separated by spaces: ")
nums= []
y=values.split()
for x in y:
    nums.append(int(x))


print("Running maximum:", running_max(nums))
````

### B3. Loops vs vectorization  *(4 marks)*

You are given `x = np.arange(1, 1_000_001)`.

1. Write a function `sum_of_squares_loop(x)` that computes the sum of squares using a **Python `for` loop**.
2. Write a function `sum_of_squares_vec(x)` that computes the same thing using **vectorized NumPy**.
3. Confirm both give the same result.
4. Use `%timeit` (or `time`) to compare their speed and briefly comment in a comment line on what you observe.

````python
# B3 — your solution here
import numpy as np
import time

# Create the array
x = np.arange(1, 1_000_001)

# Function using a Python for loop
def sum_of_squares_loop(x):
    total = 0
    for i in x:
        total += i ** 2
    return total

# Function using vectorized NumPy
def sum_of_squares_vec(x):
    return np.sum(x ** 2)

# Confirm both give the same result
result_loop = sum_of_squares_loop(x)
result_vec = sum_of_squares_vec(x)

print("Loop result      :", result_loop)
print("Vectorized result:", result_vec)
print("Same result?     :", result_loop == result_vec)

# Time the loop version
start = time.time()
sum_of_squares_loop(x)
end = time.time()
print("Loop time:", end - start, "seconds")

# Time the vectorized version
start = time.time()
sum_of_squares_vec(x)
end = time.time()
print("Vectorized time:", end - start, "seconds")

#The vectorized NumPy implementation is much faster than the Python loop version
````

3. `top_student` — the **index** of the student with the highest total.

````python

# B4 — your solution here
import numpy as np

rng = np.random.default_rng(42)
scores = rng.integers(40, 101, size=(5, 4))
print(scores)

# B4 — your solution here

# 1. Average score of each student
student_avg = np.mean(scores, axis=1)

# 2. Highest score in each subject
subject_max = np.max(scores, axis=0)

# 3. Index of the student with the highest total score
top_student = np.argmax(np.sum(scores, axis=1))

# 4. Standard deviation of all scores
overall_std = np.std(scores)

print("Student averages:", student_avg)
print("Subject maximums:", subject_max)
print("Top student index:", top_student)
print("Overall standard deviation:", overall_std)
````

### B5. Boolean masking  *(4 marks)*

Using the `scores` array from B4:

1. `passing` — a boolean array, `True` where score >= 60.
2. `num_passing` — total count of passing scores.
3. `high_scores` — a 1-D array of all scores strictly greater than 90.
4. Replace every score below 50 with 50 (a "grade floor"), storing the result in `floored` **without modifying** `scores`.

````python
# B5 — your solution here
import numpy as np

# Using the scores array from B4

# 1. Boolean array: True where score >= 60
passing = scores >= 60

# 2. Total number of passing scores
num_passing = np.sum(passing)

# 3. 1-D array of all scores greater than 90
high_scores = scores[scores > 90]

# 4. Replace scores below 50 with 50 without modifying the original array
floored = np.where(scores < 50, 50, scores)

print("Passing:")
print(passing)

print("\nNumber of passing scores:", num_passing)

print("\nHigh scores (>90):")
print(high_scores)

print("\nFloored scores:")
print(floored)
````

### B6. Reshaping & broadcasting  *(3 marks)*

1. Take `np.arange(12)` and reshape it into a 3×4 array called `grid`.
2. Create a 1-D array `col_weights = np.array([1, 2, 3, 4])` and multiply each **column** of `grid`
   by the corresponding weight using **broadcasting** (no loops). Store as `weighted`.
3. Subtract the **row mean** from each element so every row of `centered` has mean 0.

````python
# B6 — your solution here
import numpy as np

# 1. Create the grid
grid = np.arange(12).reshape(3, 4)

# 2. Apply column weights using broadcasting
col_weights = np.array([1, 2, 3, 4])
weighted = grid * col_weights

# 3. Subtract the row mean from each row
centered = weighted - weighted.mean(axis=1, keepdims=True)

print("grid:\n", grid)
print("\nweighted:\n", weighted)
print("\ncentered:\n", centered)
````

Solve the system of linear equations:

```
2x +  y -  z =  8
```

1. Build the coefficient matrix `A` and the right-hand side vector `b`.

### C1. Monte Carlo estimate of π  *(6 marks)*

You can estimate π by throwing random darts at a unit square `[0,1] x [0,1]` and counting how many
land inside the quarter circle of radius 1 centered at the origin. The fraction inside, times 4,
approximates π.

Write a function `estimate_pi(n, seed=0)` that:
1. Generates `n` random `(x, y)` points in the unit square using a seeded `np.random.default_rng`.
2. Counts how many fall inside the quarter circle (`x**2 + y**2 <= 1`) — **vectorized, no loops**.
3. Returns the estimate of π as a float.

Then call it with `n = 1_000_000` and print your estimate alongside `np.pi`.

````python
# C1 — your solution here
import numpy as np

def estimate_pi(n, seed=0):
    rng = np.random.default_rng(seed)

    # Generate n random (x, y) points in [0,1] x [0,1]
    points = rng.random((n, 2))
    x = points[:, 0]
    y = points[:, 1]

    # Count points inside the quarter circle
    inside = (x**2 + y**2) <= 1

    # Estimate π
    return 4 * np.mean(inside)

# Run the simulation
n = 1_000_000
pi_estimate = estimate_pi(n)

print("Estimated π:", pi_estimate)
print("NumPy π:     ", np.pi)
print("Error:       ", abs(pi_estimate - np.pi))
````
